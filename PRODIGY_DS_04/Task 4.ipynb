#### Analyze and visualize sentiment patterns in social media data to understand public opinion and attitude towards specific topics or brands.

```python
#Import libraries

import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```python
#Dataset
df = pd.read_csv('twitter_training.csv')
```

```python
nltk.download('vader_lexicon')
```

```python
#Sentiment Analyzer
sia = SentimentIntensityAnalyzer()
```

```python
df.columns
```

```python
# Now try the sentiment analysis again
df['sentiment_score'] = df['Borderlands'].apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
topic_sentiment = df.groupby('Borderlands')['sentiment_score'].mean()
```

```python
topic_sentiment.plot(kind='bar')
plt.title('Sentiment Analysis of Borderlands-related Tweets')
plt.xlabel('Borderlands')
plt.ylabel('Average Sentiment Score')
plt.show()
```
